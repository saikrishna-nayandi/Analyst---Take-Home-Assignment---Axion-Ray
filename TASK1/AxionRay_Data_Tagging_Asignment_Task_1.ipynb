```python
import pandas as pd
import re
df = pd.read_excel("DA - Task 1-updated.xlsx")
df.head(10)
```

### Cleaning the text by removing special charactersn from these specific columns with Regular expression

```python
def clean_text(text):
    # Remove special characters, symbols, and punctuations except alphanumeric characters and spaces
    return re.sub(r'[^A-Za-z0-9\s]', '', str(text))

# Clean all relevant columns
df['Complaint'] = df['Complaint'].apply(clean_text)
df['Cause'] = df['Cause'].apply(clean_text)
df['Correction'] = df['Correction'].apply(clean_text)
df.head(10)
```

```python
# Provided keywords list for the "root cause" column
keywords = [
    "Not Tightened", "Not Installed", "Not Mentioned", "Loosened", "Not Included",
    "Out of Fitting", "Blown", "Poor Material", "Leaking", "Failed Sending",
    "No Oring", "Not Tighten", "Out of Range", "Lubricant Drip Drown", "Fault",
    "Internal Issue", "Screwed in a Thread", "Faulty"
]

# Normalize keywords for matching
keywords_lower = [k.lower() for k in keywords]

# Function to extract root cause keywords
def extract_root_cause(row):
    combined_text = f"{row['Complaint']} {row['Cause']} {row['Correction']}".lower()
    matches = [kw for kw, kw_lower in zip(keywords, keywords_lower) if kw_lower in combined_text]
    return ', '.join(matches) if matches else 'Unknown'

# Apply function to each row
df['Root Cause'] = df.apply(extract_root_cause, axis=1)
df.head(10)
```
